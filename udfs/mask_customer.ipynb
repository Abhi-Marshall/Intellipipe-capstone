```python
# ------------------------------------------------------------------------------
# 2. Register: udf_mask_customer_id
# ------------------------------------------------------------------------------
spark.sql("""
CREATE OR REPLACE FUNCTION capstone_project.capstone_schema.udf_mask_customer_id(cust_id STRING)
RETURNS STRING
LANGUAGE PYTHON
COMMENT 'Applies SHA-256 hashing to customer_id to pseudonymise PII'
AS $$
    import hashlib
    if not cust_id: return None
    return hashlib.sha256(cust_id.encode()).hexdigest()
$$
""")

# ------------------------------------------------------------------------------
# 4. Apply Governance 
# ------------------------------------------------------------------------------
print("Applying Unity Catalog execution grants...")

spark.sql("GRANT EXECUTE ON FUNCTION capstone_project.capstone_schema.udf_mask_customer_id TO `account users`")

print("Success! UDFs are registered to capstone_project.capstone_schema")
```
