```python
# Create a secret scope and store credentials using Databricks SDK
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()

# Create the secret scope
try:
    w.secrets.create_scope(scope="intellipipe")
    print("✅ Created secret scope: intellipipe")
except Exception as e:
    print(f"Scope might already exist: {e}")

# Store your service principal credentials
w.secrets.put_secret(
    scope="intellipipe",
    key="databricks_host",
    string_value="dbc-5e2e6f3e-0d80.cloud.databricks.com"
)
w.secrets.put_secret(
    scope="intellipipe",
    key="sp_client_id",
    string_value="9973688f-291f-40c3-b252-af582c10fd84"
)
w.secrets.put_secret(
    scope="intellipipe",
    key="sp_client_secret",
    string_value="dose896dd6a4fb15efe0a10117c9e3dbd493"
)

print("✅ Secrets stored successfully!")
```

```python
# Grant READ permission to the user on the secret scope
from databricks.sdk import WorkspaceClient
from databricks.sdk.service.workspace import AclPermission

w = WorkspaceClient()

# Grant ACL permission
w.secrets.put_acl(
    scope="intellipipe",
    principal="capstoneproject196@gmail.com",
    permission=AclPermission.READ
)

print("✅ ACL permission granted successfully!")
```
