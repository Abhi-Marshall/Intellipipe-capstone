```python
# ------------------------------------------------------------------------------
# 1. Register: udf_normalize_category
# ------------------------------------------------------------------------------
spark.sql("""
CREATE OR REPLACE FUNCTION capstone_project.capstone_schema.udf_normalize_category(cat STRING)
RETURNS STRING
LANGUAGE PYTHON
COMMENT 'Standardises free-text category values using a lookup dictionary'
AS $$
    if not cat: return "Unknown"
    clean_cat = cat.strip().upper()
    category_mapping = {
        "ELECTRONICS": "Electronics", "ELECT.": "Electronics", "ELEC": "Electronics",
        "APPAREL": "Apparel", "APP": "Apparel",
        "BEAUTY": "Beauty", "BEAUTY PRODUCTS": "Beauty",
        "SPORTS": "Sports", "SPORT": "Sports"
    }
    return category_mapping.get(clean_cat, clean_cat.title())
$$
""")

# ------------------------------------------------------------------------------
# 4. Apply Governance 
# ------------------------------------------------------------------------------
print("Applying Unity Catalog execution grants...")

spark.sql("GRANT EXECUTE ON FUNCTION capstone_project.capstone_schema.udf_normalize_category TO `account users`")

print("Success! UDFs are registered to capstone_project.capstone_schema")
```
